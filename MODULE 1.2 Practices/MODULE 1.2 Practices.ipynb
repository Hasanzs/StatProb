SOAL 1

Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

## RANGE

```python
# Dataset
gol_pemain = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Hitung rentang
rentang_data = max(gol_pemain) - min(gol_pemain)

print("Rentang data:", rentang_data)
```

## HISTOGRAM RANGE

```python
import matplotlib.pyplot as plt

# Dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Membuat Histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Membuat batas koordinat
y = 1  # Koordinat Y dimana garis akan digambar
x1 = (min(data) - 1) / (max(data) - min(data))  # X mula-mula
x2 = (max(data) + 1) / (max(data) - min(data))  # X akhir

# Garis Horizontal
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {max(data) - min(data)}')

# Garis Vertikal
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')

# Membuat Label
plt.xlabel('Jumlah Gol')
plt.ylabel('Frekuensi')
plt.title('Range')

# Menambah Legenda
plt.legend()

# Menunjukkan Plot
plt.show()
```

## VARIANSI


```python
import numpy as np

# Dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Menghitung variansi menggunakan NumPy
variance = np.var(data)

print(f"Variansi: {variance:.2f}")
```

## STANDAR DEVIASI

```python
import numpy as np

# Dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Menghitung standar deviasi menggunakan NumPy
std_deviation = np.std(data)

print(f"Standar Deviasi: {std_deviation:.2f}")
```

## CONCLUSION

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]

# Menghitung statistik
range_data = max(data) - min(data)
variance_data = np.var(data)
std_deviation_data = np.std(data)

# Menyiapkan label dan nilai untuk diagram lingkaran
labels = ['Rentang', 'Variansi', 'Standar Deviasi']
values = [range_data, variance_data, std_deviation_data]

# Membuat diagram lingkaran
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
ax.axis('equal')  # Agar diagram lingkaran terlihat sebagai lingkaran

# Menambahkan judul
plt.title('Statistik Data Gol Pemain')

# Menampilkan diagram lingkaran
plt.show()
```

# SOAL 2

Dalam tim olahraga, tinggi pemain adalah: 180 cm, 185 cm, 175 cm, 190 cm, 170 cm, 195 cm, 165 cm. Bagaimana sebaran data dari tinggi pemain ini?


## RANGE

```python
# Dataset
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Hitung rentang
rentang_tinggi = max(tinggi_pemain) - min(tinggi_pemain)

print("Rentang tinggi pemain:", rentang_tinggi, "cm")
```

## HISTOGRAM RANGE

```python
import matplotlib.pyplot as plt

# Dataset
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Membuat histogram
plt.hist(tinggi_pemain, bins=10, edgecolor='black', alpha=0.7, color='blue')

# Menambahkan garis vertikal untuk rentang
plt.axvline(x=min(tinggi_pemain), color='red', linestyle='dashed', linewidth=1, label=f'Min Tinggi: {min(tinggi_pemain)} cm')
plt.axvline(x=max(tinggi_pemain), color='red', linestyle='dashed', linewidth=1, label=f'Max Tinggi: {max(tinggi_pemain)} cm')

# Membuat label
plt.xlabel('Tinggi Pemain (cm)')
plt.ylabel('Frekuensi')
plt.title('Histogram Tinggi Pemain')

# Menambahkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

## VARIANSI

```python
import numpy as np

# Dataset
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Menghitung variansi menggunakan NumPy
variansi_tinggi = np.var(tinggi_pemain)

print(f"Variansi tinggi pemain: {variansi_tinggi:.2f} cm^2")
```

## STANDAR DEVIASI

```python
import numpy as np

# Dataset
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Menghitung standar deviasi menggunakan NumPy
std_deviasi_tinggi = np.std(tinggi_pemain)

print(f"Standar Deviasi tinggi pemain: {std_deviasi_tinggi:.2f} cm")
```

## HISTOGRAM STANDAR DEVIASI

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Membuat histogram
plt.hist(tinggi_pemain, bins=10, edgecolor='black', alpha=0.7, color='orange')

# Menghitung standar deviasi
std_deviasi_tinggi = np.std(tinggi_pemain)

# Menambahkan garis vertikal untuk standar deviasi
plt.axvline(x=np.mean(tinggi_pemain), color='blue', linestyle='dashed', linewidth=1, label=f'Mean: {np.mean(tinggi_pemain):.2f} cm')
plt.axvline(x=np.mean(tinggi_pemain) - std_deviasi_tinggi, color='red', linestyle='dashed', linewidth=1, label=f'Std Dev -')
plt.axvline(x=np.mean(tinggi_pemain) + std_deviasi_tinggi, color='red', linestyle='dashed', linewidth=1, label=f'Std Dev +')

# Membuat label
plt.xlabel('Tinggi Pemain (cm)')
plt.ylabel('Frekuensi')
plt.title('Histogram Tinggi Pemain dengan Standar Deviasi')

# Menambahkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

## CONCLUSION

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Menghitung statistik
rentang_tinggi = max(tinggi_pemain) - min(tinggi_pemain)
variansi_tinggi = np.var(tinggi_pemain)
std_deviasi_tinggi = np.std(tinggi_pemain)

# Menyiapkan label dan nilai untuk diagram lingkaran
labels = ['Rentang', 'Variansi', 'Standar Deviasi']
values = [rentang_tinggi, variansi_tinggi, std_deviasi_tinggi]

# Membuat diagram lingkaran
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
ax.axis('equal')  # Agar diagram lingkaran terlihat sebagai lingkaran

# Menambahkan judul
plt.title('Statistik Data Tinggi Pemain')

# Menampilkan diagram lingkaran
plt.show()
```
