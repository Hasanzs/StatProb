Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A: 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

## MEAN

```python
import numpy as np

# Data nilai UAS
nilai_uas = np.array([95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75])

# Hitung mean
mean_nilai_uas = np.mean(nilai_uas)

# Tampilkan hasil
print("Mean nilai UAS:", mean_nilai_uas)
```

## HISTOGRAM

```python
import numpy as np
import matplotlib.pyplot as plt

# Data nilai UAS
nilai_uas = np.array([95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75])

# Hitung mean
mean_nilai_uas = np.mean(nilai_uas)

# Gambarkan histogram
plt.hist(nilai_uas, bins=20, edgecolor='black', alpha=0.7, color='blue')

# Tambahkan garis vertikal untuk mean
plt.axvline(mean_nilai_uas, color='red', linestyle='dashed', linewidth=2, label='Mean')

# Tambahkan Label
plt.xlabel('Nilai UAS')
plt.ylabel('Frekuensi')
plt.title('Histogram Nilai UAS')

# Tambahkan Legend
plt.legend()

# Tampilkan plot
plt.show()
```

## HISTOGRAM

```python
import numpy as np
import matplotlib.pyplot as plt

# Data nilai UAS
nilai_uas = np.array([95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75])

# Hitung modus
unique_values, counts = np.unique(nilai_uas, return_counts=True)
modus_values = unique_values[counts == np.max(counts)]

# Gambarkan histogram
plt.hist(nilai_uas, bins=20, edgecolor='black', alpha=0.7, color='blue')

# Tambahkan garis vertikal untuk setiap modus
for modus in modus_values:
    plt.axvline(modus, color='orange', linestyle='dashed', linewidth=2, label=f'Modus = {modus}')

# Tambahkan Label
plt.xlabel('Nilai UAS')
plt.ylabel('Frekuensi')
plt.title('Histogram Nilai UAS')

# Tambahkan Legend
plt.legend()

# Tampilkan plot
plt.show()
```
