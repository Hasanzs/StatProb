Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

## MEAN

```python
import numpy as np

tinggi_badan_kelas_A = np.array([100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100])

mean_tinggi = np.mean(tinggi_badan_kelas_A)

print("Mean tinggi badan:", mean_tinggi)
```

# HISTOGRAM

```python
import numpy as np
import matplotlib.pyplot as plt

# Data tinggi badan siswa kelas A
tinggi_badan_kelas_A = np.array([100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100])

# Hitung mean
mean_tinggi = np.mean(tinggi_badan_kelas_A)

# Gambarkan histogram
plt.hist(tinggi_badan_kelas_A, bins=20, edgecolor='black')
plt.axvline(mean_tinggi, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.title('Distribusi Tinggi Badan Siswa Kelas A')
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

# MODUS

```python
from collections import Counter
import matplotlib.pyplot as plt
import math

Data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

counter = Counter(Data)

#ubah data ke dictionary
mode_data = dict(counter)

# mencari data yang paling banyak muncul
n = max(list(counter.values()))
print(f"angka terbanyak muncul sebanyak: {n} kali")

modus = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        modus.append(k)
print(f"Modus : {modus}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

DATA = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung frekuensi kemunculan setiap nilai
counter = Counter(DATA)

# Ubah data ke dictionary
modus_data = dict(counter)

# Mencari data yang paling banyak muncul
n = max(list(counter.values()))
modus = [k for k, v in modus_data.items() if v == n]

# Gambarkan histogram
plt.hist(DATA, bins=30, edgecolor='black', alpha=0.5, color='green')

# Membuat garis lurus pada modus
for k in modus:
    plt.axvline(x=k, color='red', linestyle='dashed', linewidth=1, label=f'Mode = {k:.2f}')

# Menambah Label
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Modus')

# Menambah Legend
plt.legend()

# Menunjukkan Plot
plt.show()
```

modus tidak ada karena, semua data tersebut keluar satu kali semua.
